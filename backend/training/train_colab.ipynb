# Fine-tune CryptoBERT cho crypto sentiment

**Setup**: Colab → Runtime → **T4 GPU**.

**Input**: upload `train.csv`, `val.csv`, `test.csv` vào Drive tại `/MyDrive/crypto_sentiment/data/`.

**Output**: model fine-tuned, push lên HuggingFace Hub.

```python
# 1. CÀI ĐẶT THƯ VIỆN & KẾT NỐI DRIVE
!pip install -q transformers datasets accelerate evaluate scikit-learn
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from datasets import Dataset, DatasetDict

# 2. KHAI BÁO BIẾN CƠ BẢN
DATA_DIR = '/content/drive/MyDrive/crypto_sentiment/data'
OUT_DIR  = '/content/drive/MyDrive/crypto_sentiment/model'
LABELS = ['bearish', 'neutral', 'bullish']
label2id = {l: i for i, l in enumerate(LABELS)}
id2label = {i: l for l, i in label2id.items()}

# 3. TẢI VÀ XỬ LÝ DỮ LIỆU
def load_and_clean(split):
    df = pd.read_csv(f'{DATA_DIR}/{split}.csv').dropna(subset=['text', 'label'])
    df['label'] = df['label'].map(label2id).astype(int)
    return Dataset.from_pandas(df[['text', 'label']], preserve_index=False)

ds = DatasetDict({s: load_and_clean(s) for s in ['train', 'val', 'test']})
```

```python
import numpy as np
from scipy.special import softmax
from sklearn.metrics import f1_score, accuracy_score
from transformers import (AutoTokenizer, AutoModelForSequenceClassification,
                          DataCollatorWithPadding)

BASE = 'ElKulako/cryptobert'
tokenizer = AutoTokenizer.from_pretrained(BASE)
ds_tok = ds.map(lambda x: tokenizer(x['text'], truncation=True, max_length=256), batched=True)

model = AutoModelForSequenceClassification.from_pretrained(
    BASE, num_labels=3, id2label=id2label, label2id=label2id, ignore_mismatched_sizes=True
)

def compute_metrics(eval_pred):
    logits, labels = eval_pred.predictions, eval_pred.label_ids
    preds = np.argmax(logits, axis=-1)
    probs = softmax(logits, axis=-1)
    return {
        'accuracy': accuracy_score(labels, preds),
        'macro_f1': f1_score(labels, preds, average='macro'),
        'avg_confidence': float(probs.max(axis=-1).mean()),
    }
```

```python
from transformers import TrainingArguments, Trainer

args = TrainingArguments(
    output_dir=f'{OUT_DIR}/checkpoints',
    num_train_epochs=2,
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    weight_decay=0.01,
    eval_strategy='epoch',
    save_strategy='epoch',
    save_total_limit=1,
    load_best_model_at_end=True,
    metric_for_best_model='macro_f1',
    fp16=True,
    seed=42,
    report_to='none',
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=ds_tok['train'],
    eval_dataset=ds_tok['val'],
    data_collator=DataCollatorWithPadding(tokenizer),
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
import matplotlib.pyplot as plt

hist = trainer.state.log_history
train = [(x['step'], x['loss']) for x in hist if 'loss' in x]
eval_ = [(x['step'], x['eval_loss'], x['eval_accuracy']) for x in hist if 'eval_loss' in x]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(*zip(*train), label='train')
ax1.plot([s for s,_,_ in eval_], [l for _,l,_ in eval_], 'o-', label='eval')
ax1.set_xlabel('steps'); ax1.set_ylabel('loss'); ax1.legend(); ax1.set_title('Loss')

ax2.plot([s for s,_,_ in eval_], [a for _,_,a in eval_], 'go-')
ax2.set_xlabel('steps'); ax2.set_ylabel('accuracy'); ax2.set_title('Eval Accuracy')
plt.tight_layout(); plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from scipy.special import softmax
import seaborn as sns
import matplotlib.pyplot as plt

pred = trainer.predict(ds_tok['test'])
y_pred = pred.predictions.argmax(axis=-1)
y_true = pred.label_ids

probs = softmax(pred.predictions, axis=-1)
confidence = probs.max(axis=-1)

print(classification_report(y_true, y_pred, target_names=LABELS, digits=4))

print("=== Confidence tren tap test ===")
print(f"  Trung binh : {confidence.mean():.4f}")
print(f"  Thap nhat  : {confidence.min():.4f}")
print(f"  Cao nhat   : {confidence.max():.4f}")
print()
for i, label in id2label.items():
    mask = y_pred == i
    if mask.sum() > 0:
        print(f"  {label:8s}: avg conf = {confidence[mask].mean():.4f}  ({mask.sum()} mau)")

sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt='d', cmap='Blues',
            xticklabels=LABELS, yticklabels=LABELS)
plt.xlabel('Predicted'); plt.ylabel('True'); plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd

df_test = pd.read_csv(f'{DATA_DIR}/test.csv').dropna(subset=['text', 'label']).reset_index(drop=True)
df_test['true_label'] = [id2label[t] for t in y_true]
df_test['pred_label'] = [id2label[p] for p in y_pred]
df_test['confidence'] = confidence
df_test['correct']    = df_test['pred_label'] == df_test['true_label']

# 10 mau model kem chac nhat
print("=== 10 mau confidence thap nhat ===")
cols = ['text', 'true_label', 'pred_label', 'confidence', 'correct']
print(df_test.nsmallest(10, 'confidence')[cols].to_string(index=False))

# Accuracy theo nguong confidence
print("\n=== Accuracy theo nguong confidence ===")
for thr in [0.5, 0.6, 0.7, 0.8, 0.9]:
    m = confidence >= thr
    if m.sum() == 0:
        continue
    print(f"  conf >= {thr:.1f}: {m.sum():4d} mau  accuracy = {df_test.loc[m, 'correct'].mean():.4f}")
```
